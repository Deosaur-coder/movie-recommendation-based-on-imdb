```python
import pandas as pd 
import numpy as np 
```

```python
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies=pd.read_csv("tmdb_5000_movies.csv")
```

```python
credits.head()
```

```python
movies.head()
```

```python
print(credits.shape)
print(movies.shape)
```

```python
cred_rename=credits.rename(index=str,columns={"movie_id":"id"})
mrg_movies=movies.merge(cred_rename,on='id')
mrg_movies.shape
```

```python
mrg_movies.head()
```

```python
movies_clean=mrg_movies.drop(columns=['homepage','title_x','title_y','status','production_countries'])
```

```python
movies_clean.info()
```

```python
v=movies_clean['vote_count']
R=movies_clean['vote_average']
c=movies_clean['vote_average'].mean()
m=movies_clean['vote_count'].quantile(0.70)
```

```python
movies_clean['weigeted_averages']=((R*v)+(c*m))/(v+m)
```

```python
movies_clean.head()
```

```python
movies_clean.tail()
```

```python
movie_ranking=movies_clean.sort_values('weigeted_averages',ascending=False)
```

```python
movie_ranking
```

```python
movie_ranking[['original_title','vote_count','vote_average','weigeted_averages','popularity']]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
weighted_average=movie_ranking.sort_values('weigeted_averages',ascending=False)
plt.figure(figsize=(12,6))
graph=sns.barplot(x=weighted_average['weigeted_averages'].head(15),y=weighted_average['original_title'].head(15))
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
movie_scale=scaler.fit_transform(movies_clean[['weigeted_averages',"popularity"]])
movie_normalised=pd.DataFrame(movie_scale,columns=['weigeted_averages',"popularity"])
movie_normalised.head()
```

```python
#adding 2 new columns
movies_clean[['normlised_weight_average',"normlaised_popularity"]]=movie_normalised
```

```python
movies_clean.head()
```

```python
movies_clean["score"]=movies_clean['normlised_weight_average']*0.5 + movies_clean["normlaised_popularity"]*0.5
movie_final=movies_clean.sort_values(["score"],ascending=False)
movie_final[["original_title","score"]]
```
